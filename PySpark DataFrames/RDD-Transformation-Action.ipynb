# Creating Spark Session

```python
import os
import sys

os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
```

```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder\
.master('local[1]')\
.appName('SparkSession')\
.getOrCreate()
```

# Reading RDD

```python
rdd = spark.sparkContext.textFile("./test.txt")
```

## flatMap()

```python
rdd2 = rdd.flatMap(lambda x : x.split(" "))
```

```python
rdd2.collect()
```

## map()

```python
rdd3 = rdd2.map(lambda x : (x,1))
```

```python
rdd3.collect()
```

## reduceByKey()

```python
rdd4 = rdd3.reduceByKey(lambda x, y: x+y)
```

```python
rdd4.collect()
```

## sortByKey()

```python
rdd5 = rdd4.map(lambda x : (x[1], x[0])).sortByKey()
print(rdd5.collect())
```

## filter()

```python
rdd6 = rdd5.filter(lambda x: 'any' in x[1])
```

```python
print(rdd6.collect())
```

# RDD Actions

## Count

```python
rdd_count = rdd5.count()
print("Count :", rdd_count)
```

## First

```python
rdd_first = rdd5.first()
print("First Record :", str(rdd_first[0]), ",",str(rdd_first[1]))
```

## Max

```python
rdd_max = rdd5.max()
print("The Max element:", str(rdd_max[0]), ",", str(rdd_max[1]))
```

## Reduce

```python
total_count = rdd5.reduce(lambda a,b: (a[0]+b[0], a[1]))
print(str(total_count[0]))
```

## take

```python
data3 = rdd5.take(3)
for f in data3:
    print(str(f[0]), f[1])
```
